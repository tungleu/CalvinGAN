### Extract text from all the images in dataset

```python
import csv
#PATH TO IMAGE DATASET 
directory = "./dataset"

texts = []
f = open('text_data.txt','w')
for image_file in os.listdir(directory):
    img = cv2.imread(os.path.join(directory,image_file))
    text = pytesseract.image_to_string(img).replace("\n","").replace("\x0c","").strip()
    if text == "":
        continue
    row = text+"\n"
    f.write(row)
f.close()
```
